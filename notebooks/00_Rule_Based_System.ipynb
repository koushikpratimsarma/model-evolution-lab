# History

Before machine learning, computers could not learn from data.

Everything had to be written manually.

If we wanted a computer to detect sentiment, we wrote rules ourselves.

Example:

IF "good"
→ Positive

IF "bad"
→ Negative

These systems worked only for small problems.

As language became more complex, this approach became impossible to maintain.

This motivated researchers to look for machine learning methods.

## ❌ Limitation of Previous Approach

There is no previous model here.

Rule-Based System is the starting point of early NLP.

In this approach, humans manually write rules.

Example:

If sentence contains "good" → Positive  
If sentence contains "bad" → Negative

The computer does not learn from data.
It only follows human-written rules.

## 💡 Core Idea

Rule-Based NLP means solving language problems using manually written `if-else` rules.

The system checks whether certain words or patterns are present in the text.

For example:

- If text contains positive words → Positive
- If text contains negative words → Negative
- Otherwise → Unknown

First Hands-on Code

```python
def rule_based_sentiment(text):
    text = text.lower()
    positive_words = ["love", "amazing", "enjoy", "good"]
    negative_words = ["hate", "bad", "dislike", "terrible"]

    if any(word in text for word in positive_words):
        return "Positive"
    elif any(word in text for word in negative_words):
        return "Negative"
    else:
        return "Unknown"

test_texts = [
    "I love AI",
    "This is bad",
    "The model is powerful"
]

for text in test_texts:
    print(text, "=>", rule_based_sentiment(text))
```

## 🧠 Improving the Rules

The first version only understands:

- good
- bad

Now we will add more positive and negative words.

## Better Rule-Based System

```python
positive_words = ["good", "great", "excellent", "amazing", "love", "happy"]
negative_words = ["bad", "terrible", "poor", "hate", "sad", "worst"]

def improved_sentiment(text):
    text = text.lower()

    for word in positive_words:
        if word in text:
            return "Positive"

    for word in negative_words:
        if word in text:
            return "Negative"

    return "Unknown"


test_sentences = [
    "I love AI",
    "This result is terrible",
    "The project is amazing",
    "The movie is average"
]

for sentence in test_sentences:
    print(sentence, "=>", improved_sentiment(sentence))
```

## ⚠️ Where Rule-Based System Fails

Rule-Based systems fail when language becomes complex.

Example:

"This movie is not good"

Our system sees the word `good`, so it predicts Positive.

But the actual meaning is Negative.

```python
failure_cases = [
    "This movie is not good",
    "This food is not bad",
    "The project is good but slow",
    "The movie was excellent but boring",
    "I do not hate this"
]

for sentence in failure_cases:
    print(sentence, "=>", improved_sentiment(sentence))
```

## 🔧 Adding Negation Handling

Now we will handle simple cases like:

- not good → Negative
- not bad → Positive

This is still rule-based, but slightly smarter.

```python
def sentiment_with_negation(text):
    text = text.lower()

    if "not good" in text:
        return "Negative"

    if "not bad" in text:
        return "Positive"

    for word in positive_words:
        if word in text:
            return "Positive"

    for word in negative_words:
        if word in text:
            return "Negative"

    return "Unknown"


for sentence in failure_cases:
    print(sentence, "=>", sentiment_with_negation(sentence))
```

## 📊 Rule-Based System Flow

Text Input  
↓  
Convert to lowercase  
↓  
Check negation rules  
↓  
Check positive words  
↓  
Check negative words  
↓  
Return sentiment

## ✅ Summary

In this notebook, I learned:

- Rule-Based NLP uses manually written rules.
- It does not learn from data.
- It works for simple cases.
- It fails when language becomes complex.
- Negation, sarcasm, context, and ambiguity are difficult to handle.
- This limitation motivated Machine Learning approaches like Bag of Words.
